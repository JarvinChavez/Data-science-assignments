```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("homework-1/Iris.csv")  
```

```python
iris.info()
display(iris.head())
```

```python
print(iris.isnull().sum())
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, height=5)
plt.show()
```

```python
g = sns.FacetGrid(iris, col="Species")
g.map(sns.histplot, "PetalLengthCm")
plt.show()
```

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
plt.show()
```

```python
sns.stripplot(x="Species", y="SepalLengthCm", data=iris, jitter=True)
plt.show()
```

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
sns.stripplot(x="Species", y="SepalLengthCm", data=iris, color="black", alpha=0.5)
plt.show()
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris)
plt.show()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species")
plt.show()
```

sns.heatmap(iris.drop("Id", axis=1).corr(), annot=True, cmap="coolwarm")
plt.show()


```python
sns.histplot(iris["SepalWidthCm"], kde=True)
plt.show()
```

```python
sns.swarmplot(x="Species", y="PetalWidthCm", data=iris)
plt.show()
```
